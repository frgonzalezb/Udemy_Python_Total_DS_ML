# Scikit-Spatial

Documentación: https://scikit-spatial.readthedocs.io/en/stable/

**Descripción general (traducida de la documentación):**

`scikit-spatial` es una biblioteca de Python que proporciona objetos espaciales y cálculos entre ellos. Los objetos básicos (puntos y vectores) son subclases de `ndarray` de NumPy. Otros objetos, como líneas, planos y círculos, tienen puntos o vectores como atributos.

## 1. Sección de muestra hecha por Federico

```python
from skspatial.objects import Point
from skspatial.plotting import plot_2d, plot_3d
```

```python
punto1 = Point([1, 2])
punto2 = Point([4, 3])

plot_2d(
    punto1.plotter(c='r', s=100, label='Punto 1'),
    punto2.plotter(c='b', s=100, label='Punto 2')
);
```

```python
punto1 = Point([1, 2, 3])
punto2 = Point([4, 5, 6])

plot_3d(
    punto1.plotter(c='r', s=100, label='Punto 1'),
    punto2.plotter(c='b', s=100, label='Punto 2')
);
```

## 2. Sección de exploración personal

Intentando hacer un "mono de palo"...

```python
import matplotlib.pyplot as plt
from skspatial.objects import Circle, Line
```

```python
head: Circle = Circle([0, 5], 1)

fig, ax = plt.subplots()
head.plot_2d(ax, fill=False)
plt.axis((-10, 10, -10, 10))
plt.show()
```

```python
# debug por problemas con el kernel
# usando código de la documentación como prueba
point = Point([0, 5])
line = Line(point=[0, 0], direction=[1, 1])
point_projected = line.project_point(point)
_, ax = plot_2d(
  point.plotter(s=50),
  point_projected.plotter(c='r', s=50, zorder=3),
  line.plotter(t_2=5, c='k'),
)
limits = ax.axis('equal')
plt.show()
```
