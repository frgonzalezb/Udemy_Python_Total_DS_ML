# Gráfico Pastel (Pie Plot) con Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Datos de ejemplo
df: pd.DataFrame = pd.DataFrame({
    'Etiquetas': ['Manzanas', 'Bananas', 'Naranjas', 'Uvas'],
    'Cantidad': [30, 60, 80, 20]
})
df
```

```python
# Generar el gráfico pastel
plt.pie(df['Cantidad']);
```

```python
# Ídem, pero con etiquetas
plt.pie(df['Cantidad'], labels=df['Etiquetas']);
```

**IMPORTANTE:** Nótese que, por defecto, la orientación va en sentido opuesto al reloj, partiendo desde las 3:00 en punto.

```python
# Ídem, pero con porcentajes
plt.pie(
    df['Cantidad'],
    labels=df['Etiquetas'],
    autopct='%1.1f%%'  # string para representar los porcentajes
);
```

```python
# Ídem, pero manipulando los colores
plt.pie(
    df['Cantidad'],
    labels=df['Etiquetas'],
    autopct='%1.1f%%',
    colors=['red', 'yellow', 'orange', 'violet']
);
```

```python
# Ídem, pero con sombra
plt.pie(
    df['Cantidad'],
    labels=df['Etiquetas'],
    autopct='%1.1f%%',
    colors=['red', 'yellow', 'orange', 'violet'],
    shadow=True
);
```

```python
# Ídem, pero destacando un área (por ejemplo, las uvas)
plt.pie(
    df['Cantidad'],
    labels=df['Etiquetas'],
    autopct='%1.1f%%',
    colors=['red', 'yellow', 'orange', 'violet'],
    shadow=True,
    explode=(0, 0, 0, 0.5)
);
```

```python
# Ídem, pero agregando un título al gráfico
plt.pie(
    df['Cantidad'],
    labels=df['Etiquetas'],
    autopct='%1.1f%%',
    colors=['red', 'yellow', 'orange', 'violet'],
    shadow=True,
    explode=(0, 0, 0, 0.5)
)
plt.title('Ventas totales de frutas, septiembre 2024');
```
