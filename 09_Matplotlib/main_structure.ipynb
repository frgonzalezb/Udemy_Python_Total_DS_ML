# Estructura Principal de Matplotlib

### Glosario (anotaciones de la clase)

**Figura (figure, fig):** Elemento gráfico más grande y corresponde al lienzo o canvas que contiene al gráfico. Cabe destacar que una figura puede contener uno o muchos gráficos (axes).

**Axes:** Corresponde al gráfico propiamente dicho y a todos los elementos que lo componen (título, ejes, etiquetas, etc.).

```python
import matplotlib.pyplot as plt
```

```python
stonks: list[int] = [1, 3, 2, 5, 4, 7, 6, 9, 8, 10]
plt.plot(stonks)
plt.show()  # explicit is better than implicit, I guess...
```

```python
# Acceder a la figura para cambiar sus propiedades
# Por ejemplo, el color de la figura
plt.figure(facecolor='red')
plt.plot(stonks)
plt.show()
```

```python
# Crear una figura con 4 subplots
plt.subplots(nrows=2, ncols=2)
plt.show()
```

```python
# También se puede crear de esta forma
fig, axs = plt.subplots(nrows=3, ncols=3)
plt.show()
```

```python
# Ver los tipos de datos de fig y ax
print(type(fig))
print(type(axs))
```

```python
# Obtener la primera fila de ax
print(axs[0])
```

```python
# Obtener el primer elemento de la primera fila de ax
print(axs[0][0])
```
