# Estilo para tus Gr√°ficos con Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Datos de ejemplo
df: pd.DataFrame = pd.DataFrame({
    'Etiquetas': ['Manzanas', 'Bananas', 'Naranjas', 'Uvas'],
    'Cantidad': [30, 35, 26, 33]
})
df
```

```python
# Gr√°ficos de ejemplo
fig, axs = plt.subplots(nrows=1, ncols=2)
ax1 = axs[0]
ax2 = axs[1]
ax1.plot(df['Etiquetas'], df['Cantidad'])
ax2.pie(df['Cantidad']);
```

```python
# Asignar un estilo predefinido
plt.style.use('ggplot')

fig, axs = plt.subplots(nrows=1, ncols=2)
ax1 = axs[0]
ax2 = axs[1]
ax1.plot(df['Etiquetas'], df['Cantidad'])
ax2.pie(df['Cantidad']);
```

```python
# Mostrar todos los estilos predefinidos
print(plt.style.available)
```

```python
# Probando otro estilo
plt.style.use('Solarize_Light2')

fig, axs = plt.subplots(nrows=1, ncols=2)
ax1 = axs[0]
ax2 = axs[1]
ax1.plot(df['Etiquetas'], df['Cantidad'])
ax2.pie(df['Cantidad']);
```
