# Creación de Múltiples Gráficos con Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Crear dos gráficos
fig, axs = plt.subplots(nrows=1, ncols=2)
```

```python
# Indexar axes para desplear un gráfico distinto en cada uno
fig, axs = plt.subplots(nrows=1, ncols=2)

ax1 = axs[0]
ax2 = axs[1];
```

```python
# No se nota visualmente, pero al ver las variables...
print(ax1, ax2)
```

```python
# Desplegar gráficos distintos para ax1 y ax2
fig, axs = plt.subplots(nrows=1, ncols=2)

ax1 = axs[0]
ax2 = axs[1]

ax1.plot([1, 2, 3, 4], [1, 4, 9 ,16])
ax2.scatter([1, 2, 3, 4], [1, 4, 9 ,16]);
```

```python
# Ídem, pero con personalización
fig, axs = plt.subplots(nrows=1, ncols=2)

ax1 = axs[0]
ax2 = axs[1]

ax1.plot([1, 2, 3, 4], [1, 4, 9 ,16])
ax2.scatter([1, 2, 3, 4], [1, 4, 9 ,16]);

ax1.set_title('Lineal')
ax1.set_xlabel('Eje X')
ax1.set_ylabel('Eje Y')

ax2.set_title('Scatter')
ax2.set_xlabel('Eje X')
ax2.set_ylabel('Eje Y');
```

```python
# Ídem, pero con mejor separación entre gráficos
fig, axs = plt.subplots(nrows=1, ncols=2)

ax1 = axs[0]
ax2 = axs[1]

ax1.plot([1, 2, 3, 4], [1, 4, 9 ,16])
ax2.scatter([1, 2, 3, 4], [1, 4, 9 ,16]);

ax1.set_title('Lineal')
ax1.set_xlabel('Eje X')
ax1.set_ylabel('Eje Y')

ax2.set_title('Scatter')
ax2.set_xlabel('Eje X')
ax2.set_ylabel('Eje Y')

fig.tight_layout();
```

```python
# Ídem, pero con ajuste de padding en el espaciado
fig, axs = plt.subplots(nrows=1, ncols=2)

ax1 = axs[0]
ax2 = axs[1]

ax1.plot([1, 2, 3, 4], [1, 4, 9 ,16])
ax2.scatter([1, 2, 3, 4], [1, 4, 9 ,16]);

ax1.set_title('Lineal')
ax1.set_xlabel('Eje X')
ax1.set_ylabel('Eje Y')

ax2.set_title('Scatter')
ax2.set_xlabel('Eje X')
ax2.set_ylabel('Eje Y')

fig.tight_layout(pad=5.0);
```
