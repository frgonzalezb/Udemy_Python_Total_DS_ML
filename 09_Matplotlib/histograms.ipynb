# Histogramas en Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Cargar datos desde CSV
path: str = './data/Ventas.csv'
df = pd.read_csv(path)
df.head()
```

```python
# Obtener información general de los datos
df.info()
```

```python
# Obtener estadísticos elementales
df.describe()
```

```python
# Crear un histograma para la columna "Producto" (forma 1)
df['Producto'].hist()
```

```python
# Crear un histograma para la columna "Producto" (forma 2)
plt.hist(df['Producto']);
```

```python
# Crear un histograma basado en la columna "Total Venta"
plt.hist(df['Total Venta']);
```

```python
# Generar un histograma gaussiano basado en un conjunto de números aleatorios
numbers: np.ndarray = np.random.randn(1000)
plt.hist(numbers);
```

```python
# Ídem, pero especificando la cantidad de bins ("resolución")
numbers: np.ndarray = np.random.randn(1000)
plt.hist(numbers, bins=100);
```

```python
# Ídem, pero especificando la magnitud de alpha ("opacidad")
numbers: np.ndarray = np.random.randn(1000)
plt.hist(numbers, bins=100, alpha=0.5);
```

```python
# Ídem, pero especificando el color de los bins
numbers: np.ndarray = np.random.randn(1000)
plt.hist(numbers, bins=100, alpha=0.5, color='red');
```

```python
# Ídem, pero especificando el color de borde para los bins
numbers: np.ndarray = np.random.randn(1000)
plt.hist(numbers, bins=50, alpha=0.5, color='red', edgecolor='green');
```

```python
# Ídem, pero especificando el estilo general del gráfico
numbers: np.ndarray = np.random.randn(1000)
plt.hist(
    numbers,
    bins=50,
    alpha=0.5,
    color='red',
    edgecolor='green',
    histtype='step'
);
```

```python
# Generar varios histogramas superpuestos
x1: np.ndarray = np.random.randn(1000)
x2: np.ndarray = np.random.randn(500)
x3: np.ndarray = np.random.randn(250)

plt.hist(x1, histtype='stepfilled', alpha=0.33, bins=50, color='red')
plt.hist(x2, histtype='stepfilled', alpha=0.33, bins=50, color='green')
plt.hist(x3, histtype='stepfilled', alpha=0.33, bins=50, color='blue');
```
