# Crea Cualquier tipo de Gráficos desde Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importar datos de iluminación desde CSV
df: pd.DataFrame = pd.read_csv('./data/Iluminacion.csv')
df
```

```python
# Federico escoge Hexbin como el mejor tipo de gráfico para estos datos
help(plt.hexbin)
```

```python
# Crear el gráfico base
fig, ax = plt.subplots()
ax.hexbin(df['Latitud'], df['Longitud'], C=df['Luminosidad'], gridsize=20);
```

```python
# Ídem, pero con mejoras
fig, ax = plt.subplots()
hexbin = ax.hexbin(
    df['Latitud'],
    df['Longitud'],
    C=df['Luminosidad'],
    gridsize=50,
    cmap='inferno',
)
ax.set_facecolor('black')
ax.set_title('Distribución de iluminación en Vice City', pad=20)
ax.set_xlabel('Latitud')
ax.set_ylabel('Longitud')
plt.colorbar(hexbin, label='Nivel de luminosidad');
```
