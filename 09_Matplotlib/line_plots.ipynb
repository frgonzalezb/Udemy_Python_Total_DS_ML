# Gráficos de Línea (Line Plots) en Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Generar un objeto Figure y Axes
fig = plt.figure()
ax = plt.axes()
print(type(fig))
print(type(ax))
print(fig, ax)
```

```python
# Generar una grilla simple
plt.grid()
```

```python
# Definir información a trazar
x: list[int] = [1, 3, 2, 5, 4]
y: list[int] = [4, 5, 2, 3, 1]
```

```python
# Trazar la información
plt.plot(x, y)
```

```python
# Trazar el mismo gráfico con grilla
plt.plot(x, y)
plt.grid()  # debe estar en la misma celda de código
```

```python
# Ídem, pero con puntos y líneas sólidas rojas
plt.plot(x, y, 'ro-')
plt.grid()
```

```python
# Ídem, pero con estrellas y líneas punteadas azules
plt.plot(x, y, 'b*:')
plt.grid()
```

```python
# Ídem, pero configurando los valores mínimos y máximos para los ejes
plt.plot(x, y, 'b*:')
plt.axis((0, 6, 0, 6))
plt.grid()
```

```python
# Nunca está demás recordar la documentación
help(plt.axis)
```

### Trazar gráfico usando datos reales (desde un CSV)

```python
# Cargar datos desde el archivo CSV
path: str = './data/Lluvias_region_A.csv'
df: pd.DataFrame = pd.read_csv(path)
df.head()
```

```python
"""
(Asumiendo que todos los datos están ok...)

Trazar un gráfico de línea en el cual:
- El nombre de cada región se ubica en el eje X
- Las precipitaciones del mes de enero se ubican en el eje Y
"""
plt.plot(df['region'], df['enero'])
```

```python
# Corregir formato del gráfico para mejor visualización
plt.plot(df['region'], df['enero'], 'b*:')
plt.xticks(rotation=45)
plt.grid()
```
