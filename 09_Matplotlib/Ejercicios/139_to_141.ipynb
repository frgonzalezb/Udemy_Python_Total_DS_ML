# Gráficos de Línea (Line Plots) en Matplotlib 1
Crea un gráfico de línea simple utilizando Matplotlib con los siguientes datos:

```
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

Debes activar el grid (o parrilla) para ver las líneas de guía que nos ayuden a medir nuestros datos.

````python
import matplotlib.pyplot as plt


x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.grid()
plt.show()
````
