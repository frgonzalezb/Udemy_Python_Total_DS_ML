# Crea Cualquier tipo de Gráficos desde Matplotlib 1
Investiga en la documentación de Matplotlib y crea un gráfico de tallo (stem):

Gráfico stem

Que represente la relación entre dos conjuntos de datos `x` e `y`.

Los valores para `x` deben ser los números enteros del 1 al 10, y los valores para `y` deben ser el cuadrado de cada número en `x`.

Asegúrate de incluir títulos adecuados para el gráfico, así como para los ejes `x` e `y`.



Título del gráfico: `"Relación entre Números y sus Cuadrados"`

Título del eje x: `"Número (x)"`

Título del eje y: `"Cuadrado del número (x^2)"`

```python
import matplotlib.pyplot as plt


# data
x: list[int] = [n for n in range(1, 11)]
y: list[int] = [n ** 2 for n in x]

# plot
fig, ax = plt.subplots()
ax.stem(x, y)
ax.set_title('Relación entre Números y sus Cuadrados')
ax.set_xlabel('Número (x)')
ax.set_ylabel('Cuadrado del número (x^2)')
plt.show()
```

# Crea Cualquier tipo de Gráficos desde Matplotlib 2
Investiga en la documentación de Matplotlib y crea un gráfico apilado (`stackplot`) con los siguientes datos:

`y1` con valores `[6, 1, 4, 4, 8]`

`y2` con valores `[4, 6, 3, 5, 8]`

`y3` con valores `[7, 9, 9, 2, 7]`

El eje x debe representar el tiempo, con valores del 1 al 5 y debe ser almacenado en una variable llamada `x`

1. Asigna los siguientes colores a cada serie: `y1` rojo, `y2` azul, `y3` verde

2. Agrega una leyenda que indique qué serie corresponde a `y1`: Serie y1, `y2`: Serie y2 y `y3`: Serie y3.

3. Incluye un título al gráfico: `"Distribución acumulativa de valores por serie de tiempo"`

4. Agrega etiquetas para los ejes x: `"Tiempo"` y y: `"Valores Acumulados"`.

```python
import numpy as np
import matplotlib.pyplot as plt


# data
x: list[int] = [n for n in range(1, 6)]
y1: list[int] = [6, 1, 4, 4, 8]
y2: list[int] = [4, 6, 3, 5, 8]
y3: list[int] = [7, 9, 9, 2, 7]
y: np.ndarray = np.vstack([y1, y2, y3])

# plot
fig, ax = plt.subplots()
ax.stackplot(x, y, colors=['red', 'blue', 'green'])
ax.set_title('Distribución acumulativa de valores por serie de tiempo')
ax.set_xlabel('Tiempo')
ax.set_ylabel('Valores Acumulados')
plt.legend(['Serie y1', 'Serie y2', 'Serie y3'], loc='upper left')
plt.show()
```
