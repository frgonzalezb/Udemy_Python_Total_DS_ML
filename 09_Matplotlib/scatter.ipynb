# Gráficos Scatter en Matplotlib

También conocidos como "gráficos de dispersión".

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Generar conjuntos de datos
x: np.ndarray = np.linspace(0, 10, 30)
y: np.ndarray = np.sin(x)
print(x, y)
```

```python
# Generar un gráfico lineal (para comparar)
plt.plot(x, y);
```

```python
# Ídem, pero mostrando sólo las ubicaciones exactas de los datos
plt.plot(x, y, 'o');
```

```python
# Ídem, pero con línea punteada y color azul
plt.plot(x, y, ':ob');
```

```python
# Ídem, pero con marcadores pentágonos y línea sólida
plt.plot(x, y, '-p');
```

```python
# Ídem, pero con marcadores pentágonos, línea sólida y color rojo
plt.plot(x, y, '-p', color='red');
```

```python
# Ídem, pero con tamaño de marcador grande
plt.plot(x, y, '-p', color='red', markersize=15);
```

```python
# Ídem, pero con otro ancho de línea
plt.plot(x, y, '-p', color='red', markersize=15, linewidth=5);
```

```python
# Ídem, pero con otro color para el relleno de los marcadores
plt.plot(
    x,
    y,
    '-p',
    color='red',
    markersize=15,
    markerfacecolor='blue',
    linewidth=5
);
```

```python
# Generar un gráfico de dispersión con números (pseudo-)aleatorios
random_numbers: np.random.RandomState = np.random.RandomState(0)
markers = ['o', 'x', '+', 'v', 's', 'd']  # tipos de puntos

for marker in markers:
    plt.plot(random_numbers.rand(5), random_numbers.rand(5), marker)
```

```python
# Generar un gráfico de dispersión con el método específico scatter
plt.scatter(x, y);
```

**NOTA:** La diferencia entre `plt.plot` y `plt.scatter` es que este último posee herramientas de personalización más específicas para este tipo de gráficos.

```python
# Probar scatter con números (pseudo-)aleatorios
random_numbers: np.random.RandomState = np.random.RandomState(0)
plt.scatter(random_numbers.rand(100), random_numbers.rand(100));
```

```python
# Ídem, pero con colores aleatorios
random_numbers: np.random.RandomState = np.random.RandomState(0)
plt.scatter(
    random_numbers.rand(100),
    random_numbers.rand(100),
    c=random_numbers.rand(100)
);
```

```python
# Ídem, pero aleatorizando el tamaño de los puntos
random_numbers: np.random.RandomState = np.random.RandomState(0)
plt.scatter(
    random_numbers.rand(100),
    random_numbers.rand(100),
    c=random_numbers.rand(100),
    s=random_numbers.rand(100) * 1000
);
```

```python
# Ídem, pero con alpha
random_numbers: np.random.RandomState = np.random.RandomState(0)
plt.scatter(
    random_numbers.rand(100),
    random_numbers.rand(100),
    c=random_numbers.rand(100),
    s=random_numbers.rand(100) * 1000,
    alpha=0.3
);
```

```python
# Ídem, pero con otro mapa de colores ('viridis' es el default)
random_numbers: np.random.RandomState = np.random.RandomState(0)
plt.scatter(
    random_numbers.rand(100),
    random_numbers.rand(100),
    c=random_numbers.rand(100),
    s=random_numbers.rand(100) * 1000,
    alpha=0.3,
    cmap='coolwarm'
);
```

```python
# Ídem, pero con un colorbar de referencia (tipo "mapa de calor")
random_numbers: np.random.RandomState = np.random.RandomState(0)
plt.scatter(
    random_numbers.rand(100),
    random_numbers.rand(100),
    c=random_numbers.rand(100),
    s=random_numbers.rand(100) * 1000,
    alpha=0.3,
    cmap='coolwarm'
)
plt.colorbar();
```
