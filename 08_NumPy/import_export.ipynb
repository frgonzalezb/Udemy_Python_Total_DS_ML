# ImportaciÃ³n y ExportaciÃ³n de Datos con NumPy

```python
import os
import numpy as np
```

```python
# Cargar datos desde un archivo CSV
file_path: str = './data/medallas.csv'
array: np.ndarray = np.genfromtxt(file_path, delimiter=',', skip_header=1)
print(array)
```

```python
# Cargar datos desde un archivo CSV y rellenar los valores NaN con ceros
file_path: str = './data/medallas.csv'
array: np.ndarray = np.genfromtxt(
    file_path,
    delimiter=',',
    skip_header=1,
    filling_values=0
)
print(array)
```

```python
"""
Cargar datos desde un archivo CSV, rellenar los valores NaN con ceros y
cambiar el tipo de dato al correcto (int).
"""
file_path: str = './data/medallas.csv'
array: np.ndarray = np.genfromtxt(
    file_path,
    delimiter=',',
    skip_header=1,
    filling_values=0,
    dtype=int
)
print(array)
```

```python
# Exportar un array de NumPy a un archivo CSV
file_path: str = './data/output.csv'
np.savetxt(file_path, array, delimiter=',', fmt='%d')
print('Success!') if os.path.exists(file_path) else print('Error!')
```
