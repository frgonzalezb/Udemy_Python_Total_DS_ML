# Funciones de Nivel Inferior en Seaborn

Funciones como `relplot()`, `displot()` o `catplot()` son denominadas de "alto nivel" porque nos ofrecen una forma rápida y sencilla de crear gráficos comunes con una sola llamada de función.

Por otro lado, las de "bajo nivel" son más complejas de construir, pero dan un control más detallado sobre cosas como la estética y disposición de los gráficos (más o menos como se hacen las cosas con Matplotlib puro).

```python
import pandas as pd
import seaborn as sns
```

```python
df: pd.DataFrame = sns.load_dataset('penguins')
df
```

### Uso de `PairGrid` para representar las relaciones que hay en el dataset

```python
graph: sns.PairGrid = sns.PairGrid(df, hue='species')
graph
```

Esto genera un mosaico de gráficos vacíos. Nótese también que hay relaciones duplicadas, en donde sólo cambia la posición de las variables en los ejes.

```python
# Eliminar relaciones duplicadas con la opción corner
graph = sns.PairGrid(df, hue='species', corner=True)
graph
```

```python
# Mostrar contenido en los gráficos
graph = sns.PairGrid(df, hue='species', corner=True)
graph.map_lower(sns.kdeplot, hue=None, level=5, color='.2')
```

##### Anotaciones relevantes

- No se puede llamar a una función de alto nivel en un objeto `PairGrid`, de ahí que se usan directamente los nombres completos de los tipos de gráficos, como en `sns.kdeplot` (en vez de un parámetro `kde`).

- Los `sns.kdeplot` aquí en cada recuadro representan una estimación de densidad de kernel que muestre la distribución de las dos variables que le corresponda a cada recuadro.

- El parámetro `hue` del método `map_lower()` puede anular al tocayo definido previamente en el objeto. Al estar en `None` permite en el ejemplo que no haya distinciones por categoría.

- El parámetro `level` especifica el número de niveles de contorno, definiendo cuántas capas tiene que dibujar (como en un mapa de relieves con curvas de nivel, donde se pueden distinguir las diferentes altitudes).

![Mapa](https://upload.wikimedia.org/wikipedia/commons/thumb/f/fa/Cntr-map-1.jpg/640px-Cntr-map-1.jpg)

- El parámtro `color` define el color (d'oh!) de los gráficos utilizando un tono gris especificado como un valor de color.

```python
# Ídem, pero superponiendo gráficos de dispersión
graph = sns.PairGrid(df, hue='species', corner=True)
graph.map_lower(sns.kdeplot, hue=None, level=5, color='.2')
graph.map_lower(sns.scatterplot, marker='+')
```

```python
# Ídem, pero rellenando los subplots en diagonal sin contenido
graph = sns.PairGrid(df, hue='species', corner=True)
graph.map_lower(sns.kdeplot, hue=None, level=5, color='.2')
graph.map_lower(sns.scatterplot, marker='+')
graph.map_diag(sns.histplot, element='step', linewidth=0, kde=True)
```

```python
# Ídem, pero añadiendo una leyenda
graph = sns.PairGrid(df, hue='species', corner=True)
graph.map_lower(sns.kdeplot, hue=None, level=5, color='.2')
graph.map_lower(sns.scatterplot, marker='+')
graph.map_diag(sns.histplot, element='step', linewidth=0, kde=True)
graph.add_legend(frameon=True)
```

```python
# Ídem, pero reubicando el cuadro de leyenda
graph = sns.PairGrid(df, hue='species', corner=True)
graph.map_lower(sns.kdeplot, hue=None, level=5, color='.2')
graph.map_lower(sns.scatterplot, marker='+')
graph.map_diag(sns.histplot, element='step', linewidth=0, kde=True)
graph.add_legend(frameon=True)

coordinates: tuple[float, float] = (.61, .6)
graph.legend.set_bbox_to_anchor(coordinates)
```
