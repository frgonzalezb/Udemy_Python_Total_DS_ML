# Variables Categóricas en Seaborn

```python
import pandas as pd
import seaborn as sns
```

```python
df: pd.DataFrame = sns.load_dataset('tips')
df
```

```python
# Observar la dispersión de ganancias totales según día
sns.catplot(
    data=df,
    x='day',
    y='total_bill',
    kind='strip',
    hue='smoker'
)
```

```python
# Ídem, pero con un tipo 'swarm' para evitar solapamiento de los puntos
sns.catplot(
    data=df,
    x='day',
    y='total_bill',
    kind='swarm',
    hue='smoker'
)
```

**NOTA:** Según Fede, los `stripplots` funcionan mejor para conjuntos de datos grandes y obtener una representación general rápida, mientras que los `swarmplots`, por ser computacionalmente más intensivos, son mejores en conjuntos más pequeños y obtener visualizaciónes más precisas.

```python
# Ídem, pero con un tipo 'box'
sns.catplot(
    data=df,
    x='day',
    y='total_bill',
    kind='box',
    hue='smoker'
)
```

**Descripción:**

Las "cajas" representan los cuartiles medios del conjunto de datos, mientras que los "bigotes" (las líneas inferior y superior), representan el resto de los cuartiles: la inferior representa el primer cuartil (`0.25` o 25%), y la superior representa el último cuartil (`0.75` o 75%).

La "línea central" de las cajas representa la mediana de los datos.

Por arriba o debajo de los "bigotes" pueden haber "valores atípicos" (valores inusualmente altos o bajos), representados generalmente con círculos, diamantes, cruces u otros símbolos.


**Cómo interpretar un boxplot:** 

1. **Dispersión:** Los `boxplots` son útiles para comparar la dispersión entre diversos grupos o categorías de datos. El rango intercuartílico de las cajas refleja la dispersión de la mitad central de los datos (el 50% que está entre 0.25 y 0.75).

2. **Asimetría:** Si la mediana no se encuentra en el centro de la caja, significa que los datos están sesgados: si la mediana está más cerca del primer cuartil, los datos están sesgados hacia la derecha, mientras que si está más cerca del tercer cuartil el sesgo será hacia la izquierda.

3. **Valores atípicos:** Pueden indicarnos alta variabilidad en los datos (en el mejor de los casos) o errores en la medición (en el peor de los casos).

```python
# Ídem, pero con un tipo 'violin'
sns.catplot(
    data=df,
    x='day',
    y='total_bill',
    kind='violin',
    hue='smoker'
)
```

**Descripción:**

Los `violinplots` son similares a los `boxplots`, pero usan una figura o mancha más compleja que se asemeja a un "violín" (NOTA PERSONAL: sinceramente, me recuerda más a una lira bizantina):

Ejem... volviendo al asunto:

El ancho o grosor de la figura representa la cantidad de repeticiones que hay para los valores del Eje Y. El contorno de la figura es análogo a un `kdeplot` que ha sido rotado verticalmente y duplicado simétricamente, lo que permite mostrar la densidad de los datos en diferentes valores y nos ofrece una visión más profunda de la distribución que en un `boxplot`.

De cualquier forma, nótese que hay un pequeño y más simple `boxplot` en cada "violín" como referencia rápida.

```python
# Ídem, pero con un tipo 'point'
sns.catplot(
    data=df,
    x='day',
    y='total_bill',
    kind='point',
    hue='smoker'
)
```

**Descripción:**

Un `pointplot` se utiliza para mostrar estimaciones puntuales (d'oh!) junto a sus intervalos de confianza, utilizando marcadores de puntos.

**Interpretación del gráfico:** En el ejemplo, cada punto representa el promedio de lo que se gasta en lass facturas totales de los clientes, y las líneas conectan esos puntos a lo largo de los días de la semana (de jueves a domingo). En este caso, los fumadores gastan más dinero en promedio que los no-fumadores en los viernes, mientras que en el resto de los días medidos es a la inversa.

En cuanto a las líneas verticales que emergen de los puntos, nos dan una idea de cuán seguros podemos estar sobre esos promedios. Entre mayor longitud de línea, mayor variabilidad (menor ajuste al promedio), y viceversa.

```python
# Ídem, pero con un tipo 'bar'
sns.catplot(
    data=df,
    x='day',
    y='total_bill',
    kind='bar',
    hue='smoker'
)
```

**Descripción:**

Un `barplot` nos muestran una relación que existe entre una variable categórica y una numérica. Cada barra (o conjunto de barras de colores, como en el ejemplo) representa una categoría (los días) y la altura indica el valor numérico asociado con esa variable categórica (el total de la factura). Las líneas negras representan el intervalo de confianza (o rango de incertidumbre) para cada barra: a menor longitud, mayor seguridad (de que la barra tiene la altura correcta).
