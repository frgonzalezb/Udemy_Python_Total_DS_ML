# Introducción a Seaborn

```python
# Importar Seaborn
import seaborn as sns
```

```python
# Aplicar el tema por defecto
sns.set_theme()
```

```python
# Cargar el dataset
df = sns.load_dataset('tips')
df
```

```python
# Crear una visualización
sns.relplot(
    data=df,
    x='total_bill',
    y='tip',
    col='time',
    hue='smoker',
    style='smoker',
    size='size'
)
```
