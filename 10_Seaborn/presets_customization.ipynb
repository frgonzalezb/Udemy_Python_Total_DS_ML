# Valores Preconfigurados y Personalizados en Seaborn

```python
import pandas as pd
import seaborn as sns
```

```python
df: pd.DataFrame = sns.load_dataset('penguins')
df
```

```python
# Patrones predetermiandos
sns.relplot(
    data=df,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='body_mass_g',
)
```

```python
# Personalizar: cambiar estilo y otros elementos
sns.set_theme(
    style='darkgrid',
)

sns.relplot(
    data=df,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='body_mass_g',
    palette='crest',
    marker='x',
    s=100
)
```

```python
# Ãdem, pero personalizando ejes
graph = sns.relplot(
    data=df,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='body_mass_g',
    palette='crest',
    marker='x',
    s=100
)
graph.set_axis_labels(
    'Largo del pico (mm)',
    'Grosor del pico (mm)',
    labelpad=10
)
```

```python
# Ãdem, pero personalizando leyenda
graph = sns.relplot(
    data=df,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='body_mass_g',
    palette='crest',
    marker='x',
    s=100
)

graph.set_axis_labels(
    'Largo del pico (mm)',
    'Grosor del pico (mm)',
    labelpad=10
)

if graph.legend is not None:
    graph.legend.set_title('Masa\ncorporal (g)')
```

```python
# Ãdem, pero ajustando el tamaÃ±o de la figura
graph = sns.relplot(
    data=df,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='body_mass_g',
    palette='crest',
    marker='x',
    s=100
)

graph.set_axis_labels(
    'Largo del pico (mm)',
    'Grosor del pico (mm)',
    labelpad=10
)

if graph.legend is not None:
    graph.legend.set_title('Masa\ncorporal (g)')

graph.figure.set_size_inches(6.5, 4.5)
```
