# Relación Estadística en Seaborn

```python
import pandas as pd
import seaborn as sns
```

```python
df: pd.DataFrame = sns.load_dataset('tips')
df
```

```python
# Verificar relación entre la factura total y la propina
sns.relplot(
    data=df,
    x='total_bill',
    y='tip'
)
```

```python
"""
NOTE: Seaborn escoge automáticamente el tipo de gráfico más apropiado,
según los datos. No obstante, es posible indicar el tipo de gráfico
manualmente (aunque, para este ejemplo, no sea una buena idea). 😛
"""
sns.relplot(
    data=df,
    x='total_bill',
    y='tip',
    kind='line'
)
```

```python
# Diferenciar datos por color (hue), según otra columna ('smoker')
sns.relplot(
    data=df,
    x='total_bill',
    y='tip',
    hue='smoker'
)
```

```python
# Ídem, pero también diferenciando por estilo, según columna 'smoker'
sns.relplot(
    data=df,
    x='total_bill',
    y='tip',
    hue='smoker',
    style='smoker'
)
```

```python
# Ídem, pero también diferenciando por tamaño, según columna 'size'
sns.relplot(
    data=df,
    x='total_bill',
    y='tip',
    hue='smoker',
    style='smoker',
    size='size'
)
```

```python
# Ídem, pero también generando un gráfico aparte para diferenciar por
# otra columna ('time')
sns.relplot(
    data=df,
    x='total_bill',
    y='tip',
    hue='smoker',
    style='smoker',
    size='size',
    col='time'
)
```

```python
# Ídem, pero quitando la leyenda
sns.relplot(
    data=df,
    x='total_bill',
    y='tip',
    hue='smoker',
    style='smoker',
    size='size',
    col='time',
    legend=False
)
```
